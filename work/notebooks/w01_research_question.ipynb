## 1. My lane (or freestyle) and why

**Lane 2: Refresh / Content Opportunity Scoring**

I have selected Lane 2 because content refreshes and optimizations represent high-leverage decisions in search engine optimization. Rather than blindly writing new content, search teams can maximize the value of their existing content inventory by identifying pages that are starting to lose traction, are becoming stale, or have high striking-distance positions. Scoring these opportunities systematically allows content editors and SEO strategists to prioritize high-impact updates efficiently and sustain organic traffic with minimal wasted effort.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.

# Verify selected lane choice is present in dataset structure
import pandas as pd
df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")
print(f"Dataset has {len(df)} rows across {df['client_id'].nunique()} clients ready for Lane 2 scoring.")
```

## 2. The question: decision, action, cost of a wrong call

**What decision does your work improve?**
This work improves the recurring decision of: *"Which content pages should our team review and refresh first today/this week to protect search visibility and recover organic traffic?"*

**Who acts on it, and what do they do?**
Content editors, SEO managers, and writers act on this. They will take our ranked recommendations along with the associated reason codes, inspect the specific pages, and execute tailored updates (e.g., expanding thin content, correcting outdated facts, optimizing on-page keyword targeting, or improving internal linking structures).

**What does a wrong answer cost?**
- *False Positives (recommending a stable/growing page for refresh):* Costs wasted editor and content creator hours (typically $100-$300 in research/writing costs per page) and the opportunity cost of neglecting pages that actually needed updates.
- *False Negatives (missing a page that is actively declining):* Costs compounding organic search traffic loss, lower user engagement, and a drop in leads or direct conversions over time.

**Why does data or ML help?**
With content inventories reaching thousands of pages across many clients, manual review is impossible. Search signals are multi-dimensional, tangled (clicks, impressions, average position, and engagement rates), and shift over time. ML helps by finding non-linear patterns of decay across all signals simultaneously and ranking them accurately.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.

# Check baseline metrics for potential target definitions
down_count = (df['trend_direction'] == 'down').sum()
print(f"Total declining pages (trend_direction == 'down'): {down_count} ({down_count/len(df)*100:.2f}%)")
```

## 3. Quick look at the data (2-3 real numbers)

Based on the starter dataset (`data/raw/content_refresh_anonymized.csv`), we find:
1. **Total Portfolio Scale:** There are **30,000 pages** across **32 distinct clients**, proving why automated scoring is necessary.
2. **Critical Decay Rate:** **16,262 pages (54.21%)** are classified as actively declining (`trend_direction == 'down'`), showing that decay is a widespread, high-priority problem.
3. **Stale/Stagnant Inventory:** There are **174 pages** that haven't been updated in 180+ days (`days_since_last_update >= 180`), indicating clear low-hanging fruit for refresh initiatives.
4. **Striking Distance Position:** The median average position is **11.4** (for active pages), placing the typical page right on the edge of the first page of Google—the perfect candidate for optimization to yield quick traffic gains.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.

total_rows = len(df)
unique_clients = df['client_id'].nunique()
declining_count = (df['trend_direction'] == 'down').sum()
declining_pct = (declining_count / total_rows) * 100
stale_count = (df['days_since_last_update'] >= 180).sum()
median_age = df['content_age_days'].median()
median_pos = df[df['avg_position'] > 0]['avg_position'].median()

print(f"Total pages in dataset: {total_rows}")
print(f"Unique clients: {unique_clients}")
print(f"Declining pages: {declining_count} ({declining_pct:.2f}%)")
print(f"Stale pages (>= 180 days since update): {stale_count}")
print(f"Median content age (days): {median_age:.1f}")
print(f"Median average position (excluding zero): {median_pos:.1f}")
```
